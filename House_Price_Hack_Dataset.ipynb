```python
# For Preprocessing Combine Both Train and test into single dataframe
# For Row Concatenation number of Columns and sequence of columns must be exactly same
```

```python
# Add Dependent Variable to test data & fill it with word test
House_Test['SalePrice']="test"
```

```python
# Row Concatenation - Top rows train data & Bottom rows test data
combined_data = pd.concat([House_Train,House_Test],axis=0) # Axis = 0 is Row Cooncatenation
```

```python
combined_data.info()
```

```python
# Split Data into Numeric & Object Cols
numcols = combined_data.select_dtypes(include=np.number)
objcols = combined_data.select_dtypes(include='object')
```

```python
numcols.columns
```

```python
# Check for Missing Value
objcols.isnull().sum().sort_values(ascending=False)
```

```python
catcols=numcols[['OverallQual','OverallCond', 'YearBuilt', 'YearRemodAdd','GarageYrBlt','MoSold', 'YrSold']]
```

```python
numcols=numcols.drop(['OverallQual','OverallCond', 'YearBuilt', 'YearRemodAdd','GarageYrBlt','MoSold', 'YrSold'],axis=1)
catcols.isnull().sum().sort_values(ascending=False)
```

```python
# Impute Missing Value with word "Not Available"
for col in objcols.columns:
  objcols[col]=objcols[col].fillna("Not Available")
```

```python
numcols.isnull().sum().sort_values(ascending=False)
```

```python
# Impute with Median
for col in numcols.columns:
  numcols[col] = numcols[col].fillna(numcols[col].median())
```

```python
catcols.isnull().sum()
for col in catcols.columns:
  catcols[col] = catcols[col].fillna(catcols[col].value_counts().idxmax())
```

```python
numcols.head()
```

```python
# Standard Scaler on numcols
```

```python
from sklearn.preprocessing import StandardScaler
numcols_scaled = StandardScaler().fit_transform(numcols)
numcols_scaled = pd.DataFrame(numcols_scaled,columns=numcols.columns)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
objcols=objcols.drop('SalePrice',axis=1)
```

```python
objcols_encoder=objcols.apply(LabelEncoder().fit_transform)
```

```python
catcols_encode=catcols.apply(LabelEncoder().fit_transform)
```

```python
catcols_encode=catcols_encode.reset_index()
```

```python
objcols_encoder=objcols_encoder.reset_index()
```

```python
# Concatenate all 3 dataframes into one dataframes
combined_data_final=pd.concat([numcols_scaled,objcols_encoder,catcols_encode],axis=1)
```

```python
# Split Data into train & test
House_Train = combined_data_final[combined_data_final['SalePrice'] !="test"]
House_Test = combined_data_final[combined_data_final['SalePrice'] =="test"]
```
